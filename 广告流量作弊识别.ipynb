## 查看一周的流量情况

```python
# data.groupby(by='V2')['V4'].count().tolist()
```

```python
y = data.groupby(by='V2')['V4'].count()
x = range(1,8)
_x_ticks = ["第{}天".format(i) for i in x]
fig = plt.figure(figsize=(20,8),dpi=80)
plt.xticks(x,_x_ticks)
plt.plot(x,y)
plt.grid(alpha = 0.9)
plt.show()
```

## IP的前两段或前三段相同

```python
data['ip_two_same'] = data['V4'].str.extract("([0-9]{1,3}['.'][0-9]{1,3}['.'])")
data['ip_three_same'] = data['V4'].str.extract("([0-9]{1,3}['.'][0-9]{1,3}['.'][0-9]{1,3}['.'])")
```

## 一周内 IP的 访问量

```python
y = data.groupby('V4')['V4'].count().sort_values(ascending=False)[:24*7]
x = range(24*7)
print(len(x))
print(len(y))
_x_ticks = y.index
fig = plt.figure(figsize=(20,8),dpi=80)
plt.xticks(x,_x_ticks,rotation=90)
plt.plot(x,y)
plt.grid(alpha = 0.9)
plt.show()
```
